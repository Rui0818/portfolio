```python
#Group the businesses by state
query = """
Select state, count(*) as bus_count
From yelp.business
Group by state
Order by bus_count DESC
"""
df = pd.read_sql(query, db)
df
```

```python
#Group the restaurants by state
query = """
Select state, count(*) as bus_count
From yelp.business
Where categories like '%Restaurants%'
Group by state
Order by bus_count DESC
"""
df = pd.read_sql(query, db)
df
```
