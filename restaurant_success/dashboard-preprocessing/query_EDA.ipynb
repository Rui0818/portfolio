Number of restaurants.

```python
query = """
Select count(*) 
From yelp.business
Where categories like '%Restaurants%'
"""
df = pd.read_sql(query, db)
df
```

How many restaurants per rating.

```python
query = """
Select stars, count(*) 
From yelp.business
Where categories like '%Restaurants%'
group by stars
"""
df = pd.read_sql(query, db)
df
```

```python
plt.figure(figsize=(8,8))
sns.barplot(data=df,x='stars',y='count(*)',color='#AA9868')
plt.title('Number of Restaurants per Rating')
```

Number of restaurants in each state

```python
query = """
Select state, count(*) 
From yelp.business
Where categories like '%Restaurants%'
Group by state
"""
df = pd.read_sql(query, db)
df
```

```python
plt.figure(figsize=(8,8))
sns.barplot(data=df,x='state',y='count(*)',color='#AA9868')
plt.title('Number of Restaurants for Each State')
```

Join business and tips on restaurants' business id.

```python
query = """
Select 
    b.name
    , b.stars
    , t.*
    
From yelp.tip t
left join yelp.business b
on t.business_id = b.business_id
Where b.categories like '%Restaurants%'
limit 10
"""
df = pd.read_sql(query, db)
df
```

```python
query = """
Select 
    count(distinct t.business_id)
From yelp.tip t
"""
df = pd.read_sql(query, db)
df
```

How many restaurants have tips records.

```python
query = """
Select 
    count(distinct t.business_id)
From yelp.tip t
 join yelp.business b
on t.business_id = b.business_id
Where b.categories like '%Restaurants%'
"""
df = pd.read_sql(query, db)
df
```
