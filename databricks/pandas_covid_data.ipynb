Import the data for all available days.

```
import glob    

path = '/dbfs/databricks-datasets/COVID/CSSEGISandData/csse_covid_19_data/csse_covid_19_daily_reports'
all_files = glob.glob(path + '/*.csv')    # all_files is a list contains all the file names

df1 = []
for filename in all_files:
  temp_df = pd.read_csv(filename)
  temp_df.columns = [c.replace('/', '_') for c in temp_df.columns]
  temp_df.columns = [c.replace(' ', '_') for c in temp_df.columns]
  
  month, day, year = filename.split('/')[-1].replace('.csv', '').split('-')    # month, day, year is a list contains number of month, day and year, data type: string
  d = datetime.date(int(year), int(month), int(day))    # for datetime.date(), it will raise TypeError if the argument is not an integer, so change the data type of year, month and day to int
  temp_df['Date'] = d
  df1.append(temp_df)
  
all_days_df = pd.concat(df1, axis=0, ignore_index=True, sort=False)   
all_days_df = all_days_df.drop(['Latitude', 'Longitude', 'Lat', 'Long_', 'FIPS', 'Combined_Key', 'Last_Update'], axis=1)
```

```
all_days_df.head()
```

How has the disease spread over time by state?

```
 # I noticed that some values occurred in the 'Province_state' column but they are not state, like 'Chicago','Chicago, IL' etc. So I set a filter to drop those invalid value. the `state` below describes the valid US state names 
state = ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California',
       'Colorado', 'Connecticut', 'Delaware', 'Diamond Princess',
       'District of Columbia', 'Florida', 'Georgia', 'Grand Princess',
       'Guam', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas',
       'Kentucky', 'Louisiana', 'Maine', 'Maryland', 'Massachusetts',
       'Michigan', 'Minnesota', 'Mississippi', 'Missouri', 'Montana',
       'Nebraska', 'Nevada', 'New Hampshire', 'New Jersey', 'New Mexico',
       'New York', 'North Carolina', 'North Dakota',
       'Northern Mariana Islands', 'Ohio', 'Oklahoma', 'Oregon',
       'Pennsylvania', 'Puerto Rico', 'Recovered', 'Rhode Island',
       'South Carolina', 'South Dakota', 'Tennessee', 'Texas', 'Utah',
       'Vermont', 'Virgin Islands', 'Virginia', 'Washington',
       'West Virginia', 'Wisconsin', 'Wyoming']
us_df = all_days_df[all_days_df['Country_Region'] == 'US']       
valid_state_df = us_df[us_df['Province_State'].isin(state)]   
by_state = valid_state_df.groupby(['Province_State','Date'])['Confirmed'].sum()
by_state
```

```
### There are several states in the same picture
# by_state.plot(title = "Confirmed Cases over Time by State", rot = 45, figsize=(20,10))
# display()
```

```
def state_plot(state):
  #this function shows confirmed cases over time in specific state
  state_df = valid_state_df[valid_state_df['Province_State']==state]   
  state_df.groupby('Date')['Confirmed'].sum().plot(title = "Confirmed Cases over Time in "+ state, rot = 45)    
  display()
#test case 1
state_plot('Virginia')
```

```
#test case 2
state_plot('California')
```

Break this down by types of cases by region.

```
by_region = all_days_df.groupby(['Country_Region','Date'])['Confirmed','Deaths','Recovered'].sum()
by_region
```

```
def region_plot(region):
  region_df = all_days_df[all_days_df['Country_Region']==region]
  region_df.groupby('Date')['Confirmed','Deaths','Recovered'].sum().plot(title='Confirmed, Deaths, Recovered over Time in '+region)
  display()
# test case 1
region_plot('US')
```

```
# test case 2
region_plot('China')
```

Wrap this up in a function and run it yourself!

```
def county_plot(country,state,county):
  county_df = all_days_df[(all_days_df['Country_Region']==country) & (all_days_df['Province_State']==state )& (all_days_df['Admin2']==county)]
  county_df.groupby('Date')['Confirmed','Deaths','Recovered'].sum().plot(title="Confirmed, Deaths, Recovered over Time in "+county, rot=45)
  display()
## test case 1
county_plot('US','Virginia','Fairfax')
```

```
## test case 2
county_plot('US','California','San Francisco')
```
